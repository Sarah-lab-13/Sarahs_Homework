```python
word_list = text.split()
print(len(word_list))
word_list_small = word_list[:500]
```

```python
#a
import string

word_list_small = [x.lower() for x in word_list_small]

word_list_set = set(word_list_small)
for word in word_list_set:
    print(f"{word}: {word_list_small.count(word)} times")
```

```python
#or
word_list_count = []
for word in word_list_set:
    word_list_count.append(word)
    word_list_count.append(word_list.count(word))
    
print(word_list_count)    
```

```python
#c
import re

string = ' '.join(word_list_small)
vowels = re.findall(r"([aeiouy]+)", string)
vowels_unique = set(vowels)
for i in vowels_unique:
    number = vowels.count(i)
    print(f"{i} times {number}")
  
```
